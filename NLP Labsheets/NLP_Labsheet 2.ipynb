## Exercise-3:Bigram frequencies of Jane Austen Novels

```python
#B.
```

```python
with open("austen-emma.txt") as fi:
    txt=fi.read()
    print(txt)
```

```python
#C1
```

```python
tokenizer=nltk.tokenize.WhitespaceTokenizer()
a_toks=tokenizer.tokenize(txt.lower())
print(a_toks)
```

```python
#C2
```

```python
a_tokfd=FreqDist(a_toks)
a_tokfd
```

```python
#C3
```

```python
a_bigrams=list(nltk.bigrams(a_toks))
a_bigrams
```

```python
#C4
```

```python
a_bigramfd=nltk.FreqDist(a_bigrams)
a_bigramfd
```

```python
from nltk.probability import ConditionalFreqDist
from nltk.tokenize import word_tokenize
```

```python
a_bigramcfd=nltk.ConditionalFreqDist(a_bigrams)
a_bigramcfd['a']
```

```python
a_bilist=[]
for i,j in a_bigrams:
        a_bilist.append(a_bigramcfd[i])
a_bilist
```

```python
#D.
```

```python
import pickle
fl=open("austen_bigramcfd.pkl",'wb')
pickle.dump(a_bilist,fl)
```
